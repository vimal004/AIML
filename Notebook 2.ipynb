```python
def tensor_converter(filepath):
    img = tf.io.read_file(filepath)
    tensor = tf.image.decode_image(img, channels=3)
    tensor = tf.image.convert_image_dtype(tensor, dtype=tf.float32)
    tensor = tf.image.resize(tensor, (224, 224))
    return tensor
```
